```python
# Temporary path setup for this notebook session
from pathlib import Path
import sys

def _find_repo_root(start: Path) -> Path:
    # Walk upwards until we find a marker that looks like the repo root
    for candidate in (start, *start.parents):
        if (candidate / "requirements.txt").exists() and (candidate / "main.py").exists():
            return candidate
    return start

repo_root = _find_repo_root(Path.cwd().resolve())
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

print("Repo root added to sys.path:", repo_root)
```
